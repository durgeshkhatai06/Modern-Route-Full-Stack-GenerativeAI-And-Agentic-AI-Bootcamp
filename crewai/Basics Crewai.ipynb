```python
import os
from pathlib import Path
from dotenv import load_dotenv
from openai import OpenAI

# Load values from crewai/.env when running locally.
load_dotenv(dotenv_path=Path.cwd() / '.env')

api_key = os.environ.get('GROQ_API_KEY')

try:
  from google.colab import userdata
  api_key = api_key or userdata.get('GROQ_API_KEY') or userdata.get('groq_api_key')
except ModuleNotFoundError:
  pass

if api_key:
  os.environ["GROQ_API_KEY"] = api_key
  print("GROQ_API_KEY loaded.")
else:
  print("GROQ_API_KEY not found. Set it in .env, your environment, or Colab userdata.")

################## OPENAI API CHECK ############################


client = OpenAI(
    base_url="https://api.groq.com/openai/v1", # This tells the code to go to Groq
    api_key=os.environ.get("GROQ_API_KEY"),
)

response = client.responses.create(
    model="llama-3.3-70b-versatile",
    instructions="You are a coding assistant that talks like a pirate.",
    input="How do I check if a Python object is an instance of a class?",
)

print(response.output_text)
```
