##### Using CrewAI

```python
# Import Agent, Crew, Task, Process from crewai for this step.
from crewai import Agent, Crew, Task, Process


# Explain the purpose of: class YourCrewName:
class YourCrewName:
    # Define the `agent_one` function.
    def agent_one(self) -> Agent:
        # Return the computed value from this function.
        return Agent(
            # Store the result of this step in `role`.
            role="Data Analyst",
            # Store the result of this step in `goal`.
            goal="Analyze data trends in the market",
            # Store the result of this step in `backstory`.
            backstory="An experienced data analyst with a background in economics",
            # Store the result of this step in `verbose`.
            verbose=True,
        # Close the current expression.
        )

    # Define the `agent_two` function.
    def agent_two(self) -> Agent:
        # Return the computed value from this function.
        return Agent(
            # Store the result of this step in `role`.
            role="Market Researcher",
            # Store the result of this step in `goal`.
            goal="Gather information on market dynamics",
            # Store the result of this step in `backstory`.
            backstory="A diligent researcher with a keen eye for detail",
            # Store the result of this step in `verbose`.
            verbose=True,
        # Close the current expression.
        )

    # Define the `task_one` function.
    def task_one(self) -> Task:
        # Return the computed value from this function.
        return Task(
            # Store the result of this step in `name`.
            name="Collect Data Task",
            # Store the result of this step in `description`.
            description="Collect recent market data and identify trends.",
            # Store the result of this step in `expected_output`.
            expected_output="A report summarizing key trends in the market.",
            # Store the result of this step in `agent`.
            agent=self.agent_one(),
        # Close the current expression.
        )

    # Define the `task_two` function.
    def task_two(self) -> Task:
        # Return the computed value from this function.
        return Task(
            # Store the result of this step in `name`.
            name="Market Research Task",
            # Store the result of this step in `description`.
            description="Research factors affecting market dynamics.",
            # Store the result of this step in `expected_output`.
            expected_output="An analysis of factors influencing the market.",
            # Store the result of this step in `agent`.
            agent=self.agent_two(),
        # Close the current expression.
        )

    # Define the `crew` function.
    def crew(self) -> Crew:
        # Return the computed value from this function.
        return Crew(
            # Store the result of this step in `agents`.
            agents=[self.agent_one(), self.agent_two()],
            # Store the result of this step in `tasks`.
            tasks=[self.task_one(), self.task_two()],
            # Store the result of this step in `process`.
            process=Process.sequential,
            # Store the result of this step in `verbose`.
            verbose=True,
        # Close the current expression.
        )
```

```python
# Import track_crewai from opik.integrations.crewai for this step.
from opik.integrations.crewai import track_crewai

# Explain the purpose of: track_crewai(project_name="crewai-opik-integration")
track_crewai(project_name="crewai-opik-integration")

# Store the result of this step in `my_crew`.
my_crew = YourCrewName().crew()
# Store the result of this step in `result`.
result = my_crew.kickoff()

# Display this information in the notebook output.
print(result)
```
